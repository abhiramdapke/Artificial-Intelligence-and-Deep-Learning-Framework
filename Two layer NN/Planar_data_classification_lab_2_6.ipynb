```python
import numpy as np
import matplotlib.pyplot as plt
from testCases import *
import sklearn
import sklearn.datasets
import sklearn.linear_model
from planar_utils import plot_decision_boundary, sigmoid, load_planar_dataset, load_extra_datasets
import import_ipynb
import Planar_data_classification_lab_2_4 as lab4
import import_ipynb
from load_dataset import*
import h5py
```

```python

def load_dataset(database_path):
    # open dataset
    dataset_db = h5py.File(database_path, "r")

    datasets = {}
    for dataset in ["train", "dev", "test"]:
        # load the train set feautres (picuture)
        datasets[dataset] = {'X': np.array(dataset_db[dataset + "_img"][:]),  # dataset features
                             'Y': np.array(dataset_db[dataset + "_labels"][:])  # dataset labels
                             }
    return datasets
```

```python
# Datasets
noisy_circles, noisy_moons, blobs, gaussian_quantiles, no_structure = load_extra_datasets()

datasets = {"noisy_circles": noisy_circles,
            "noisy_moons": noisy_moons,
            "blobs": blobs,
            "gaussian_quantiles": gaussian_quantiles}
```

```python
noisy_circles= datasets['noisy_circles']
X,Y = noisy_circles

Y_new= np.reshape(Y,(Y.shape[0],1))
Y2 =np.reshape( Y_new,(1,Y.shape[0]))
X = np.reshape(X,(X.shape[1],X.shape[0]))
print(X.shape,Y2.shape)
plt.scatter(X[0, :], X[1, :], c=Y2.ravel(),cmap=plt.cm.Spectral);
```

```python
parameters = lab4.nn_model(X,Y2,n_h = 10, num_iterations = 10000, print_cost=True)
```

```python
predictions = lab4.predict(parameters, X)
print ('Accuracy: %f' % float((np.dot(Y2,predictions.T) + np.dot(1-Y2,1-predictions.T))/float(Y2.size)*100) + '%')
```

```python
plot_decision_boundary(lambda x: lab4.predict(parameters, x.T),X,Y2.ravel())
plt.title("Decision Boundary for hidden layer size " + str(10))
```
